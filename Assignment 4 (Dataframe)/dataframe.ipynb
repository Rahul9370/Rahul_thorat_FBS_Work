```python
import numpy as np
import pandas as pd
```

```python
customer = {
    'Order ID':[1001,1002,1003,1004,1005,1006,1007,1008],
    'Customer':['Alice','Bob','Charlie','Alice','David','Charlie','Bob','Alice'],
    'Product':['Laptop','Phone','Chair','Table','Laptop','Phone','Chair','Phone'],
    'Category':['Electronics','Electronics','Furniture','Furniture','Electronics','Electronics','Furniture','Electronics'],
    'Quantity':[2,1,4,1,1,2,2,1],
    'Price':[800,500,100,250,800,500,100,500],
    'Total Sales':[1600,500,400,250,800,1000,200,500]
}

# 1. Create a DataFrame:
# o Load the above dataset into a Pandas DataFrame.

customer = pd.DataFrame(customer)
customer
```

```python
# 2. Add a New Column:
# o Add a column Discount that gives a 10% discount on total
# sales.
customer['Discount'] = customer['Total Sales']*0.10
customer
```

```python
# 3. Add a New Row:
# o Add a new sales record: Order ID = 1009, Customer =
# "Eve", Product = "Tablet", Category = "Electronics",
# Quantity = 1, Price = 300, Total Sales = 300.

customer.loc[8] = [1009,'Eve','Tablet','Electronics',1,300,300,30.0]
customer
```

```python
# 4. Filter Data:
# o Retrieve only the sales records where the Category is
# "Electronics".
Electronics_sales = customer[customer['Category']=="Electronics"]
Electronics_sales
```

```python
# 5. Selection:
# o Select the Customer and Total Sales columns.
cust_sale = customer[['Customer','Total Sales']]
cust_sale
```

```python
# 6. Grouping:
# o Group the data by Customer and calculate the total sales per customer.

total_sale_per_cust = customer.groupby('Customer')['Total Sales'].sum()
total_sale_per_cust
```

```python
# 7. Sorting:
# o Sort the DataFrame based on Total Sales in descending order.
customer.sort_values('Total Sales',ascending=False)
```

```python
# 8. Apply Function:
# o Apply a lambda function to create a new column Sales
# Category, where values are labeled as "High" if Total Sales >
# 1000, otherwise "Low".

customer['Sales Category'] = customer['Total Sales'].apply(lambda x: 'High' if x > 1000 else 'Low')
customer
```

```python
# 9. Save DataFrame:
# o Save the final DataFrame as a CSV file named sales_data.csv.
customer.to_csv('sales_data.csv', index=False)
sales_data = customer.copy()
sales_data
```
